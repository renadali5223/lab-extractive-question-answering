```python
from datasets import load_dataset

df = load_dataset("squad", split="train").to_pandas()
df = df[['title', 'context', 'question']]
df = df.drop_duplicates(subset='context')
df.head()
```

```python
import getpass

OPENAI_API_KEY = getpass.getpass("Enter your OpenAI API key: ")
PINECONE_API_KEY = getpass.getpass("Enter your Pinecone API key: ")
```

```python
from pinecone import Pinecone, ServerlessSpec

spec = ServerlessSpec(cloud="aws", region="us-east-1")

pc = Pinecone(api_key=PINECONE_API_KEY)

index_name = "question-answering"

if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=spec
    )

index = pc.Index(index_name)
```

```python
import torch
from sentence_transformers import SentenceTransformer

device = 'cuda' if torch.cuda.is_available() else 'cpu'
retriever = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1', device=device)
```

```python
from tqdm.auto import tqdm

batch_size = 64

for i in tqdm(range(0, len(df), batch_size)):
    end = min(i + batch_size, len(df))
    batch = df.iloc[i:end]

    emb = retriever.encode(batch["context"].tolist()).tolist()
    meta = batch.to_dict(orient="records")
    ids = [f"id-{i+j}" for j in range(end - i)]

    to_upsert = list(zip(ids, emb, meta))
    _ = index.upsert(vectors=to_upsert)
```

```python
from transformers import pipeline

model_name = 'deepset/electra-base-squad2'
reader = pipeline(tokenizer=model_name, model=model_name, task='question-answering', device=0 if device=='cuda' else -1)
```

```python
def get_context(question, top_k):
    xq = retriever.encode([question]).tolist()
    xc = index.query(vector=xq[0], top_k=top_k, include_metadata=True)
    return [match['metadata']['context'] for match in xc['matches']]

from pprint import pprint

def extract_answer(question, context):
    results = []
    for c in context:
        answer = reader(question=question, context=c)
        answer["context"] = c
        results.append(answer)
    sorted_result = sorted(results, key=lambda x: x['score'], reverse=True)
    pprint(sorted_result)
    return sorted_result
```

```python
question = "How much oil is Egypt producing in a day?"
context = get_context(question, top_k=1)
extract_answer(question, context)

question = "What are the first names of the men that invented youtube?"
context = get_context(question, top_k=1)
extract_answer(question, context)

question = "What is Albert Eistein famous for?"
context = get_context(question, top_k=1)
extract_answer(question, context)

question = "Who was the first person to step foot on the moon?"
context = get_context(question, top_k=3)
extract_answer(question, context)
```
